```python
from pyspark.sql import functions as f
from pyspark.sql.functions import col, lit, when, make_date
```

```python
data=[[2021,10,30],
      [2021,12,3],
      [2022,1,14],
      [None,None,None],
      [2024,2,29],
      [2024,2,29],
      [2025,2,28],
      [2022,5,24],
      [2023,3,21],
      [2023,7,18],
      [2023,12,None]
]

column=["year","month","day"]

df_components=spark.createDataFrame(data,column)
display(df_components)
```

```python
df_date= df_components\
    .withColumn("date",f.make_date(f.col("year"),f.col("month"),f.col("day")))

display(df_date)
```

```python
from pyspark.sql.functions import lit
spark.range(1)\
    .select("id",make_date(lit(2026),lit(1),lit(1)).alias("new_year"),
            make_date(lit(2026),lit(1),lit(None)).alias("new_year_null")).show(truncate=False)
```

```python
data=[(2022,7,25),(2023,1,1),(2022,13,31),(2023,2,29),(2024,5,19),(2025,8,29)]

df_invalid= spark.createDataFrame(data,["year","month","day"])

```

```python
spark.conf.set("spark.sql.ansi.enabled",False)
df_invalid_final= df_invalid\
        .withColumn("date",f.make_date(col("year"),col("month"),col("day")))

display(df_invalid_final)
```

```python
data=[(2022,7,25),(2023,1,1),(2022,13,31),(2024,5,19),(2025,8,29)]

df_error=spark.createDataFrame(data,["year","month","day"])


display(df_error)
```

```python
df_error_final= df_error\
    .withColumn("date",when((f.col("month")>=1) & (f.col("month")<=12) & (f.col("day")>=1) & (f.col("day")<=31),f.make_date(f.col("year"),f.col("month"),f.col("day"))).otherwise(lit(None)))


display(df_error_final)
```
