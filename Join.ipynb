```python
import pyspark.sql.functions as sf
order_list=[("01", "02", 1, 350),
               ("01", "04", 1, 580),
               ("01", "07", 2, 320),
               ("02", "03", 1, 450),
               ("02", "06", 1, 220),
               ("03", "01", 1, 195),
               ("04", "09", 1, 270),
               ("04", "08", 2, 410),
               ("05", "02", 1, 350)]

order_df=spark.createDataFrame(order_list).toDF("order_id","prod_id","qty","unit_price")
```

```python
product_list= [("01", "Scroll Mouse", 250, 20),
                    ("02", "Optical Mouse", 350, 20),
                    ("03", "Wireless Mouse", 450, 50),
                    ("04", "Wireless Keyboard", 580, 50),
                    ("05", "Standard Keyboard", 360, 10),
                    ("06", "16 GB Flash Storage", 240, 100),
                    ("07", "32 GB Flash Storage", 320, 50),
                    ("08", "64 GB Flash Storage", 430, 25)]


product_df= spark.createDataFrame(product_list).toDF("prod_id","prod_name","list_price","qty")
```

```python
join_expr= order_df.prod_id==product_df.prod_id

join_df= order_df.join(product_df,join_expr,"left")\
    .drop(product_df.prod_id)\
        .withColumn("stock_qty",product_df.qty)\
            .drop(product_df.qty)\
                .withColumn("prod_name",sf.coalesce(product_df.prod_name,order_df.prod_id))


display(join_df)
```

```python
display(join_df.orderBy("order_id"))
```

```python
join_df3=join_df.selectExpr("order_id","prod_name","qty","unit_price","list_price","(list_price - unit_price) * qty as discount", "qty * unit_price as amount")\
    .orderBy(sf.desc("order_id"))


display(join_df3)
```
