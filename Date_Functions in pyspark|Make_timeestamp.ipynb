```python
from pyspark.sql.functions import col, lit, make_timestamp
```

```python
data=[(2018,2,14,10,0,45.887,'CET')]
schema=["year","month","day","hour","minute","second","tz"]

df_ts=spark.createDataFrame(data,schema)

display(df_ts)
```

```python
df_ts_final=df_ts\
    .withColumn("timestamp_col",make_timestamp(col("year"),col("month"),col("day"),col("hour"),col("minute"),col("second"),col("tz")))

display(df_ts_final)
```

```python
from pyspark.sql.functions import to_timestamp, concat, lit


df=spark.range(1)\
    .select(lit("2014-12-28").alias("date"),
            lit("06:30:45").alias("time"))
    
df=df\
    .withColumn("datetime_str",concat(df.date,lit(" "),df.time))

display(df)


df_final=df\
    .withColumn("timestamp",to_timestamp(df.datetime_str,"yyyy-MM-dd HH:mm:ss"))

display(df_final)
```

```python
from pyspark.sql import functions as f
from pyspark.sql.types import StructType,StructField,IntegerType,StringType,DoubleType

data=[
    (2019,2,13,10,30,0,'UTC'),
    (2018,2,14,10,0,45.887,"CET"),
    (2022,12,31,14,45,20,"UTC"),
        (2020,10,2,18,0,0,"America/Los_Angeles"),
    (2020,10,2,18,0,None,"America/Los_Angeles")]

schema=StructType([StructField("year",IntegerType()),
    StructField("month",IntegerType()),
    StructField("day",IntegerType()),
    StructField("hour",IntegerType()),
    StructField("minute",IntegerType()),
    StructField("second",IntegerType()),
    StructField("tz",StringType())])

df_ts=spark.createDataFrame(data,schema)

display(df_ts)
```

```python
df_ts_final=df_ts\
    .withColumn("timestamp_col",f.make_timestamp(col("year"),col("month"),col("day"),col("hour"),col("minute"),col("second"),col("tz")))


display(df_ts_final)
```

```python
from pyspark.sql.functions import lit , make_timestamp

df_lit= spark.createDataFrame([(1,)],["id"])

df_lit=df_lit\
    .withColumn("new_year_midnight",make_timestamp(lit(2026),lit(1),lit(1),lit(0),lit(0),lit(0)))\
        .withColumn("new_year_mifnight_null"\
            ,make_timestamp(lit(2026),lit(1),lit(1),lit(0),lit(0),lit(None)))


display(df_lit)
```

```python
data= [(2022,1,1,10,30,0),
       (2023,1,1,14,45,0),
       (2022,13,31,23,59,59)]


columnss= ["year_col","month_col","day_col","hour_col","minute_col","second_col"]

df_error=spark.createDataFrame(data,columnss)

display(df_error)
```

```python
from pyspark.sql.functions import col,when

df_error_final= df_error\
    .withColumn("timestamp_col",
                when((col("month_col")>=1) & (col("month_col")<=12),
        make_timestamp(col("year_col"),col("month_col"),col("day_col"),col("hour_col"),col("minute_col"),col("second_col"))).otherwise(lit(None)))

display(df_error_final)
        
```
